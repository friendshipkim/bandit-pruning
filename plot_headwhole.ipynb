```python
matplotlib.rc('font', size=10) #, weight='bold')
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14)
```

```python
fig = plt.figure(figsize=(7, 4), dpi=100)
ax = fig.add_subplot(111)
N = len(UCB1)
## necessary variables
ind = np.arange(N)                # the x locations for the groups

# bandits
plt.plot(ind , UCB1 , color="blue", linewidth=2.5, linestyle="-", label="UCB1")
plt.plot(ind , EpsilonGreedy, color="red", linewidth=2.5, linestyle="-", label="Epsilon Greedy")
plt.plot(ind , AnnealingEpsilonGreedy, color="cyan", linewidth=2.5, linestyle="-", label="Annealing Epsilon Greedy")
plt.plot(ind , Softmax, color="purple", linewidth=2.5, linestyle="-", label="Softmax")
plt.plot(ind , AnnealingSoftmax, color="yellow", linewidth=2.5, linestyle="-", label="Annealing Softmax")
plt.plot(ind , Hedge, color="green", linewidth=2.5, linestyle="-", label="Hedge")
plt.plot(ind , Exp3, color="grey", linewidth=2.5, linestyle="-", label="Exp3")

# random
plt.plot(ind , Random, color="magenta", linewidth=2.5, linestyle="-", label="Random")

plt.legend(loc = 3)

# ax.set_ylim([0.74, 0.83])
ax.set_yticks(np.arange(0.3, 0.85, 0.05))

plt.xlabel('Total number of pruned heads', fontweight='bold', fontsize='14')
plt.ylabel('Accuracy after pruning', fontweight='bold', fontsize='14')
plt.grid(True)
plt.show()
```

## print number

```python
for prune_ratio in np.arange(0.1, 1.1, 0.1):
    idx = int(72*prune_ratio - 1)
    print('--------prune {}%----------'.format(int(prune_ratio*100)))
    print('UCB: %.2f' % (UCB1[idx]*100))
    print('EpsilonGreedy: %.2f' % (EpsilonGreedy[idx]*100))
    print('AnnealingEpsilonGreedy: %.2f' % (AnnealingEpsilonGreedy[idx]*100))
    print('Softmax: %.2f' % (Softmax[idx]*100))
    print('AnnealingSoftmax: %.2f' % (AnnealingSoftmax[idx]*100))
    print('Hedge: %.2f' % (Hedge[idx]*100))
    print('Exp3: %.2f' % (Exp3[idx]*100))
    print('Random: %.2f' % (Random[idx]*100))

    print('---------------------------')
```

## ablation

```python
pruning_type = 'head_whole_ablation'
```

```python
algo_name = 'AnnealingSoftmax'
```

```python
fig = plt.figure(figsize=(7, 4), dpi=100)
ax = fig.add_subplot(111)
N = 12*6
## necessary variables
ind = np.arange(N)                # the x locations for the groups

color_list = ['#092333', '#3F349F', '#8A528C', '#D56B6C', '#F4A53C', 'yellow', 'red', 'green', 'blue', 'purple']
for thres in np.arange(0.1, 0.6, 0.1):
    label = "%.1f" % thres
    result_path = os.path.join('./results', pruning_type, algo_name, label)
    prune_after_accuracy = np.load(os.path.join(result_path, "prune_after_accuracy.npy"))
    plt.plot(ind , prune_after_accuracy , color=color_list.pop(0), linewidth=2.5, linestyle="-", label=label)

plt.legend(loc = 3)
# plt.legend(ncol=8, bbox_to_anchor=(1, 3))
plt.xlabel('Number of heads pruned', fontweight='bold', fontsize='14')
plt.ylabel('Accuracy after pruning', fontweight='bold', fontsize='14')
# ax.set_xticks(ind)
# ax.set_ylim([0.74, 0.83])
# ax.set_yticks(np.arange(0.74, 0.83, 0.01))
plt.grid(True)

plt.show()
```

## no thres vs. thres

```python
pruning_type = 'head_whole_nothres'
```

```python
algo_list = ["UCB1",
             "EpsilonGreedy",
             "AnnealingEpsilonGreedy",
             "Softmax",
             "AnnealingSoftmax",
             "Hedge",
             "Exp3",]
```

```python
algo_name = algo_list[0]
algo_name
```

```python
fig = plt.figure(figsize=(7, 4), dpi=100)
ax = fig.add_subplot(111)
N = 12*6

ind = np.arange(N)         

nothres_result_path = os.path.join('./results', pruning_type, algo_name)
thres_result_path = os.path.join('./results/head_whole', algo_name)

nothres_prune_after_accuracy = np.load(os.path.join(nothres_result_path, "prune_after_accuracy.npy"))
thres_prune_after_accuracy = np.load(os.path.join(thres_result_path, "prune_after_accuracy.npy"))

plt.plot(ind , nothres_prune_after_accuracy , color='red', linewidth=2.5, linestyle="-", label='nothres')
plt.plot(ind , thres_prune_after_accuracy , color='green', linewidth=2.5, linestyle="-", label='thres')

plt.title(algo_name, fontweight='bold', fontsize='14')
plt.legend(loc = 3, fontsize='14')
plt.xlabel('Number of heads pruned', fontweight='bold', fontsize='14')
plt.ylabel('Accuracy after pruning', fontweight='bold', fontsize='14')
ax.set_xticks(np.arange(0, 80, 10))

plt.show()
```
